```python
# Out of pipeline
# Save the resized image for easier reuse

#Training Data Set
for i in range(0, 6):
    print(i)
    path_of_the_directory='../data/men/'+str(i)+'/'
    path_of_the_directory_result='resized_imgs/men/'+str(i)+'/'
    isExist = os.path.exists(path_of_the_directory_result)
    if not isExist:
        os.makedirs(path_of_the_directory_result)
        print("The new directory is created!")
    else:
        shutil.rmtree(path_of_the_directory_result, ignore_errors=False, onerror=None)
        os.mkdir(path_of_the_directory_result)
    k = 0
    for filename in os.listdir(path_of_the_directory):
        if(filename[-3:]=='JPG'):
            cv2.imwrite(path_of_the_directory_result+filename,train_images[str(i)][k])
            k += 1

    path_of_the_directory='../data/women/'+str(i)+'/'
    path_of_the_directory_result='resized_imgs/women/'+str(i)+'/'
    isExist = os.path.exists(path_of_the_directory_result)
    if not isExist:
        os.makedirs(path_of_the_directory_result)
        print("The new directory is created!")
    else:
        shutil.rmtree(path_of_the_directory_result, ignore_errors=False, onerror=None)
        os.mkdir(path_of_the_directory_result)
    k = 0
    for filename in os.listdir(path_of_the_directory):
        if(filename[-3:]=='JPG'):
            cv2.imwrite(path_of_the_directory_result+filename,train_images[str(i)][k])
            k += 1
```

```python
# Step(2) Preprocess the images

```

```python
# Step(3) Extract features
X_train = []
Y_train = []

for i in range(6):
    print(i)
    for img in train_images[str(i)]:
            img = resize(img, (64*4, 128*4))
            fd, hog_image = hog(img, orientations=9, pixels_per_cell=(8, 8), cells_per_block=(2, 2), visualize=True)
            X_train.append(fd)
            Y_train.append(i)
print('Features Extracted')
```

```python
# Step(4) Train our classifier

# Create an SVM classifier with a linear kernel
clf = svm.SVC(kernel='rbf', C=1, degree=3)
# Fit the classifier to the training data
clf.fit(X_train, Y_train)
print('Model Trained')
```

```python
# Step(5) Save our model
filename = "../test_pipeline/Trained_SVM.joblib"
joblib.dump(clf, filename)
print('Model exported')
print('Done')
```
